# Day 18 - Turtle Graphics & Hirst Painting Project

```python
# Turtle Graphics challenges from 3 - 5

from turtle import Turtle, Screen
import random

# Create turtle and screen
screen = Screen()
screen.setup(width=800, height=800)  # Larger canvas
screen.bgcolor("white")  
screen.colormode(255)  # Enable RGB colors

# Configure turtle
myturtle = Turtle()
myturtle.shape("classic")
myturtle.pensize(1)
geometry = 360

def random_colour():
    r = random.randint(0, 255)
    g = random.randint(0, 255)
    b = random.randint(0, 255)
    myturtle.pencolor((r, g, b))

def challenge_3():
    screen.title("Draw different shape")
    for shape in range(3, 11): 
        random_colour()
        corners = geometry / shape  
        for _ in range(shape): 
            myturtle.forward(50)
            myturtle.right(corners)

def challenge_4():
    screen.title("Random Walk")
    myturtle.pensize(10)
    myturtle.speed(0)  # Maximum speed
    
    # Better movement function with more directions and cleaner code
    def random_walk():
        # Options for movement
        directions = [0, 90, 180, 270]  # All four directions
        myturtle.setheading(random.choice(directions))
        myturtle.forward(30)
        random_colour()
    
    # Draw random walk
    for _ in range(200):
        random_walk()
        
        # Keep turtle on screen by checking position
        x, y = myturtle.position()
        if abs(x) > 380 or abs(y) > 380:
            myturtle.penup()
            myturtle.goto(0, 0)
            myturtle.pendown()

def challenge_5():
    screen.title("Spirograph")
    myturtle.speed(0)
    
    for _ in range(36): 
        random_colour()
        myturtle.circle(100)
        myturtle.setheading(myturtle.heading() + 10) 

# Get user choice with proper string comparison
challenge = input("Which challenge do you want to run (3, 4 or 5)? ")
if challenge == "3":
    challenge_3()
elif challenge == "4":
    challenge_4()
elif challenge == "5":
    challenge_5()    
else:
    print("Invalid choice. Please enter 3 or 4.")

# Only call exitonclick once at the end
screen.exitonclick()
```

```python
# Hirst Painting Project

import colorgram

colors = colorgram.extract("/Users/iray/Downloads/image.jpg", 19)
color_rgb = []

for color in colors:
    r = color.rgb.r
    g = color.rgb.g
    b = color.rgb.b
    new_color = (r, g, b)
    color_rgb.append(new_color)
```

```python
from turtle import Turtle, Screen
import random

# Create turtle and screen
screen = Screen()
screen.title("Hirst Painting Project")
screen.setup(width=400, height=400)  # Larger canvas
screen.bgcolor("white")  
screen.colormode(255)  # Enable RGB colors

painter = Turtle()
painter.penup()  # This prevents drawing lines when moving
painter.hideturtle()
painter.speed(0)

start_x = -100
start_y = -100
painter.goto(start_x, start_y)


def painting():
    y_position = start_y
    for _ in range(10):
        painter.goto(start_x, y_position)
        for _ in range(10):
            painter.dot(10, random.choice(color_rgb))
            painter.forward(20)
            
        y_position += 20

painting()

screen.exitonclick()
```
